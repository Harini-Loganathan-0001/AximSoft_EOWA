```python
outlier_list = []
for col in continuous_cols:
    Q1 = house_data[col].quantile(0.25)
    Q3 = house_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = house_data[(house_data[col] < lower) | (house_data[col] > upper)]
    outlier_list.append([col, len(outliers)])

outlier_table = pd.DataFrame(outlier_list, columns=["Feature","Outlier Count"])
outlier_table
```

```python
Q1 = house_data['SalePrice'].quantile(0.25)
Q3 = house_data['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

saleprice_outliers = house_data[
    (house_data['SalePrice'] < lower) |
    (house_data['SalePrice'] > upper)]

saleprice_outliers[['GrLivArea', 'OverallQual', 'LotArea', 'SalePrice']]
```

## Feature Engineering

```python
# 1. Total square footage
house_data['TotalSF'] = (house_data['TotalBsmtSF'] + house_data['1stFlrSF'] + house_data['2ndFlrSF'])
```

```python
# 2. Total bathrooms
house_data['TotalBathrooms'] = (house_data['FullBath'] +(0.5 * house_data['HalfBath']) +
    house_data['BsmtFullBath'] + (0.5 * house_data['BsmtHalfBath']))
```

```python
# 3. House Age
house_data['HouseAge'] = house_data['YrSold'] - house_data['YearBuilt']
```

```python
# 4. RemodelAge
house_data['RemodelAge'] = house_data['YrSold'] - house_data['YearRemodAdd']
```

```python
# 5. Living space
house_data['LivingSpaceEfficiency'] = (house_data['GrLivArea'] / house_data['LotArea'])
```

```python
# 6. Total Porch Area

house_data['TotalPorchSF'] = (house_data['OpenPorchSF'] + house_data['EnclosedPorch'] +
    house_data['3SsnPorch'] + house_data['ScreenPorch'])
```

```python
house_data.shape
```

Sixe new features were created using existing variables to better represent the characteristics of each house. These features include Total Square Footage, Total Bathrooms, House Age, Remodel Age, Living space efficiency and Total Porch Area.

```python
results = []
n = X_test.shape[0]
p = house_data_2.shape[1] - 2
for name, model in models.items():

    y_prediction = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_prediction)
    mse = mean_squared_error(y_test, y_prediction)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_prediction)
    adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))
    results.append({
        "Model":name,
        "MAE":mae,
        "MSE":mse,
        "RMSE":rmse,
        "R2 Score":r2,
        "Adjusted R2":adjusted_r2
    })

results = pd.DataFrame(results)
results
```

```python
# cross validation score
results = []

for name, model in models.items():

    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    cv = cross_val_score(
        model,
        X,
        y,
        cv=5,
        scoring="r2",
        n_jobs=-1
    ).mean()

    adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))

    results.append({
        "Model": name,
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R2 Score": r2,
        "Adjusted R2": adjusted_r2,
        "CV Score": cv
    })

results = pd.DataFrame(results)
results = results.sort_values(by="R2 Score", ascending=False)
results
```
